## applymap():  Executes a function for each element in the dataFrame.

```python
# example 1
def makeCapitalLetters(x):
    return x.upper()

data = {"name": 
        ["Sophia","Gozie","Jerimiah","Immanuel",
         "Paul","Dappa","Godwin","Ruth"],
        "country":
        ["Usa","Canada","Germany","Spain",
         "France","Poland","Russia","Italy"]
       }

df = pd.DataFrame(data)

newdf = df.applymap(makeCapitalLetters)
newdf
```

```python
# example 1
def squareValues(x):
    return x*x

bioData = {"Age": 
        [24,30,18,26,29,40,14,19],
        "Children":
        [2,1,1,3,4,5,0,1],
        "Weight":
        [77,80,35,56,61,70,45,49]
       }

df = pd.DataFrame(bioData)

df1 = df.applymap(squareValues)

df1
```

## One-Hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
## One-Hot encoding is used to convert categorical variables into
# a format that can be readily used by machine learning algorithms.
# we create new variables that uses variables of 1 and 0 to represent the original variables.
```

```python
# DataFrame originally created
footballLeague= pd.DataFrame({"Teams": 
                              ['A','A','B','B','B','B','C','C','C'],
                             "Points": 
                              [25,12,15,14,19,23,25,29,17]
                             })
footballLeague
```

```python
# process of creating instance of one-hot-encoder
encoder = OneHotEncoder(handle_unknown='ignore')
# performing one-hot-encoder on Teams column
encoder_footballLeague = pd.DataFrame(encoder.fit_transform(footballLeague[['Teams']]).toarray())
# merge one-hot-encoding column back to dataframe
final_footballLeague = footballLeague.join(encoder_footballLeague)
# drop team column                                  
final_footballLeague.drop('Teams', axis=1, inplace=True)
# rename column
final_footballLeague.columns = ['Points','Team A','Team B','Team C']

final_footballLeague
```
